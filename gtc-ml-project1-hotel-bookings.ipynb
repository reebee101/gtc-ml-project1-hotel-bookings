```python
cols = ['adr','lead_time','stays_in_weekend_nights','stays_in_week_nights']

plt.figure(figsize=(12,8))
for i, col in enumerate(cols, 1):
  plt.subplot(2,2,i)
  sns.boxplot(x=df[col])
  plt.title('boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
for col in cols:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
  
  print(f"\nColumn: {col}")
  print(f"Lower bound: {lower_bound}, Upper bound: {upper_bound}")
  print(f"Number of outliers: {len(outliers)}")
  print(outliers.head(2))
  
```

```python
'''
The dataset contains a lot of missing values, especially the company column as well as
having a lot of outliers in numerical columns, so even though the dataset is big
it may not be the best considering the missing values
'''
```

```python
df['company'] = df['company'].fillna('none')
df['agent'] = df['agent'].fillna('none')

mode_country = df['country'].mode()[0]
df['country'] = df['country'].fillna(mode_country)

median_children = df['children'].median()
df['children'] = df['children'].fillna(median_children)
df['children'] = df['children'].astype(int)

df.isnull().sum()
```

```python
duplicates_count = df.duplicated().sum()
print(f'number of duplicates {duplicates_count}')

df = df.drop_duplicates()
df = df.reset_index(drop=True)
```

```python
df['adr'] = df['adr'].clip(upper=1000)
df['lead_time'] = df['lead_time'].clip(upper=365)
df['stays_in_weekend_nights'] = df['stays_in_weekend_nights'].clip(upper=5)
df['stays_in_week_nights'] = df['stays_in_week_nights'].clip(upper=7)

```

```python
cols_to_plot = ["adr", "lead_time", "stays_in_weekend_nights", "stays_in_week_nights"]

plt.figure(figsize=(14, 6))
for i, col in enumerate(cols_to_plot, 1):
    plt.subplot(2,2, i)
    sns.boxplot(y=df[col], color="skyblue")
    plt.title(f"Boxplot of {col} (after capping)")
plt.tight_layout()
plt.show()
```

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], errors='coerce')

df['arrival_date'] = pd.to_datetime(
    df['arrival_date_year'].astype(str) + '-' +
    df['arrival_date_month'].astype(str) + '-' +
    df['arrival_date_day_of_month'].astype(str),
    format='%Y-%B-%d',
    errors='coerce'
)
df.drop(['arrival_date_year','arrival_date_month','arrival_date_day_of_month'],axis=1)

cols = list(df.columns)
cols.remove('arrival_date')

insert_at = cols.index('arrival_date_week_number') + 1
cols.insert(insert_at, 'arrival_date')
```

```python
df['total_guests'] = df['adults'] + df['children'] + df['babies']

df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']

df['is_family'] = ((df['children'] > 0) | (df['babies'] > 0)).astype(int)

df.head()
```

```python
low_cardinality_cols = ['meal', 'market_segment', 'distribution_channel', 
                        'reserved_room_type', 'deposit_type', 'customer_type']

df = pd.get_dummies(df, columns=low_cardinality_cols, drop_first=True)
```

```python
country_counts = df['country'].value_counts(normalize=True)

rare_countries = country_counts[country_counts < 0.01].index
df['country_grouped'] = df['country'].replace(rare_countries, 'Other')
```

```python
df = df.drop(['reservation_status', 'reservation_status_date','reservation_changes', 
              'days_in_waiting_list'], axis=1)
```
